```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize, Normalizer

n = 200
t = np.arange(n)
x = np.random.normal(loc=1, scale=2, size=n)

plt.figure(figsize=(10,3))
plt.plot(t,x, label="x", color="orange")
plt.legend()
plt.show()
```

Zad. 1 Dla przykładowego szeregu czasowego rozwiąż zadanie normalizacji korzystając z
dostępnych w języku Python narzędzi. Przygotuj odpowiedni kod w tym języku. Zilustruj
efekty normalizacji przykładowego szeregu czasowego. 

```python
x = x.reshape(-1, 1)

x_normalized = normalize(x, norm='l2')

plt.figure(figsize=(10,3))
plt.plot(t, x_normalized, label='normalized', color="black")
plt.plot(t, x, label='orginal', color="orange")

plt.legend()
plt.show()

print(np.max(x_normalized))
print(np.min(x_normalized))
```

Zad. 2 Wygeneruj szereg czasowy. Korzystając z narzędzi dostępnych w języku Python,
przygotuj kod który dokona standaryzacji wygenerowanego szeregu czasowego. Efekty
standaryzacji zilustruj odpowiednimi wykresami.


```python
x = np.random.normal(loc=1, scale=3, size=n)
x = x.reshape(-1, 1)
     

scaler = StandardScaler()
standarized = scaler.fit_transform(x)

plt.figure(figsize=(10,3))
plt.plot(t, standarized, label='standarized', color="black")
plt.plot(t, x, label='original', color="orange")

plt.legend()
plt.show()

print(np.mean(standarized))
print(np.std(standarized))
     
```
