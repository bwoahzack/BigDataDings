```python
import numpy as np
import pandas as pd
import random
import nolds
import matplotlib.pyplot as plt
from sklearn import preprocessing
from scipy.stats import kurtosis


N = 100
t = np.arange(N)
time_range = pd.date_range(start='2023-01-01', end='2023-04-10')
y = np.random.normal(loc=2, scale=2, size=N)

plt.figure(figsize=(10, 3))
plt.plot(time_range, y, color="black")
plt.show()
```

```python
print(f"Mean: {np.mean(y)}")
print(f"Standard deviation: {np.std(y)}")
print(f"Max value: {np.max(y)}")
print(f"Min value: {np.min(y)}")
print(f"Median: {np.median(y)}")
print(f"Kurtosis: {kurtosis(y)}")
```

Wykładnik Hursta jest używany jako miara pamięci długoterminowej szeregów czasowych. Odnosi się on do autokorelacji szeregu czasowego i tempa, w jakim maleją one wraz ze wzrostem opóźnienia między parami wartości. Badania dotyczące wykładnika Hursta zostały pierwotnie opracowane w hydrologii w celu praktycznego określenia optymalnej wielkości tamy dla zmiennych warunków deszczu i suszy na rzece Nil, które były obserwowane przez długi okres czasu.[1][2] Nazwa "wykładnik Hursta" lub "współczynnik Hursta" pochodzi od Harolda Edwina Hursta (1880-1978), który był głównym badaczem w tych badaniach; użycie standardowego zapisu H dla współczynnika również odnosi się do jego nazwiska.

W geometrii fraktalnej uogólniony wykładnik Hursta został oznaczony przez H lub Hq na cześć zarówno Harolda Edwina Hursta, jak i Ludwiga Otto Höldera (1859-1937) przez Benoît Mandelbrota (1924-2010).[3] H jest bezpośrednio związany z wymiarem fraktalnym, D, i jest miarą "łagodnej" lub "dzikiej" losowości serii danych.[4]

Wykładnik Hursta jest określany jako "indeks zależności" lub "indeks zależności dalekiego zasięgu". Określa on względną tendencję szeregu czasowego do silnej regresji do średniej lub do grupowania się w danym kierunku.[5] Wartość H w zakresie 0,5-1 wskazuje na szereg czasowy z długoterminową dodatnią autokorelacją, co oznacza, że po wysokiej wartości w szeregu prawdopodobnie nastąpi kolejna wysoka wartość i że wartości w długim czasie w przyszłości również będą miały tendencję do bycia wysokimi. Wartość w zakresie 0-0,5 wskazuje na szereg czasowy z długoterminowym przełączaniem między wysokimi i niskimi wartościami w sąsiadujących parach, co oznacza, że po pojedynczej wysokiej wartości prawdopodobnie nastąpi niska wartość, a wartość po niej będzie miała tendencję do bycia wysoką, przy czym ta tendencja do przełączania między wysokimi i niskimi wartościami będzie trwała przez długi czas w przyszłości. Wartość H=0,5 może wskazywać na całkowicie nieskorelowany szereg, ale w rzeczywistości[według kogo?] jest to wartość stosowana do szeregów, dla których autokorelacje przy małych opóźnieniach czasowych mogą być dodatnie lub ujemne, ale gdzie wartości bezwzględne autokorelacji rozkładają się wykładniczo szybko do zera. Jest to przeciwieństwo typowo potęgowego rozkładu dla przypadków 0,5 < H < 1 i 0 < H < 0,5.
