```python
import numpy as np
import pandas as pd
import random
import nolds
import matplotlib.pyplot as plt
from sklearn import preprocessing
from scipy.stats import kurtosis


N = 100
t = np.arange(N)
time_range = pd.date_range(start='2023-01-01', end='2023-04-10')
y = np.random.normal(loc=2, scale=2, size=N)

plt.figure(figsize=(10, 3))
plt.plot(time_range, y, color="black")
plt.show()
```

```python
print(f"Mean: {np.mean(y)}")
print(f"Standard deviation: {np.std(y)}")
print(f"Max value: {np.max(y)}")
print(f"Min value: {np.min(y)}")
print(f"Median: {np.median(y)}")
print(f"Kurtosis: {kurtosis(y)}")
```

```python
print(f"Entropy: {nolds.sampen(y)}")
print(f"Fractal dimension: {nolds.corr_dim(y, emb_dim=1)}")
print(f"Hurst exponent: {nolds.hurst_rs(y)}")
```

Badania dotyczce wykadnika Hursta zostay pierwotnie opracowane w hydrologii w celu praktycznego okrelenia optymalnej wielkoci tamy dla zmiennych warunkw deszczu i suszy na rzece Nil, ktre byy obserwowane przez dugi okres czasu.[1][2] Nazwa "wykadnik Hursta" lub "wspczynnik Hursta" pochodzi od Harolda Edwina Hursta (1880-1978), ktry by gwnym badaczem w tych badaniach; uycie standardowego zapisu H dla wspczynnika rwnie odnosi si do jego nazwiska.

W geometrii fraktalnej uoglniony wykadnik Hursta zosta oznaczony przez H lub Hq na cze zarwno Harolda Edwina Hursta, jak i Ludwiga Otto Hldera (1859-1937) przez Benot Mandelbrota (1924-2010).[3] H jest bezporednio zwizany z wymiarem fraktalnym, D, i jest miar "agodnej" lub "dzikiej" losowoci serii danych.[4]

Wykadnik Hursta jest okrelany jako "indeks zalenoci" lub "indeks zalenoci dalekiego zasigu".
